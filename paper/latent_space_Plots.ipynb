```python
def plot_paper_energies(energies1, energies2, binwidth=2.2, beta=None, save_image=False):
    """
    Plot the energies of the samples produced by the histograms   
    UPDATE: bin now found using bin boundaries
    """
    fig, ax = plt.subplots(figsize=(60, 30))
    data = np.concatenate((energies1,energies2), axis=0)
    bins =  np.arange(min(data), max(data) + binwidth, binwidth)
    ax.hist(energies1, bins=bins, label = "QPU Samples", edgecolor='black', linewidth=8, color = "white",  histtype='stepfilled', density=True)  
    ax.hist(energies2, bins=bins, label = "Classical Samples", color = 'black', alpha=0.2, density=True) 
    plt.legend(loc='upper right', fontsize=90, framealpha=0)
    if (beta!=None):
        plt.title(r'$\beta_{eff}^{*} = $'+format(beta,'.3f'), fontsize = 60)
    ax.set_xlabel("Energy", fontsize=120)
    ax.set_ylabel("Solution Probabilities", fontsize=120, labelpad=30)

    for spine in ax.spines.values():
        spine.set_color('black') 
        spine.set_linewidth(8)
    
    ax.tick_params(axis='both', which='major', labelsize=110)
    ax.grid(True)
    plt. grid(False)
#     if (save_image==False): 
#         plt.show()
    plt.savefig("energy_histogram.pdf")
    plt.show()
    plt.close()
    #return fig
```

```python
plot_paper_energies(dwave_energies[0], aux_crbm_energy_exps)
```
